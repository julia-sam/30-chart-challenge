```python
import requests
import pandas as pd

# Step 1: Search for the Ngram and get IDs
def search_ngram(query, corpus='eng', flags='cs'):
    url = f"https://api.ngrams.dev/{corpus}/search"
    params = {
        'query': query,
        'flags': flags
    }
    response = requests.get(url, params=params)
    if response.status_code != 200:
        raise Exception(f"Error fetching data: {response.status_code} - {response.text}")
    
    data = response.json()
    ngram_ids = [ngram['id'] for ngram in data.get('ngrams', [])]
    return ngram_ids

# Step 2: Fetch year-based match counts using the ID
def fetch_ngram_by_id(ngram_id, corpus='eng'):
    url = f"https://api.ngrams.dev/{corpus}/{ngram_id}"
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Error fetching ngram details: {response.status_code} - {response.text}")
    
    return response.json()

# Aggregating the data
def aggregate_ngram_data(ngram_ids, corpus):
    all_data = pd.DataFrame()
    for ngram_id in ngram_ids:
        details = fetch_ngram_by_id(ngram_id, corpus)
        if 'stats' in details:
            df = pd.DataFrame(details['stats'])
            df['year'] = df['year'].astype(int)
            df.set_index('year', inplace=True)
            all_data = all_data.add(df, fill_value=0) if not all_data.empty else df
    return all_data

try:
    # Step 1: Search for the term and get its ngram IDs
    german_ids = search_ngram("Internet", 'ger', 'cs')
    english_ids = search_ngram("Internet", 'eng', 'cs')
    russian_ids = search_ngram("Интернет", 'rus', 'cs')

    # Step 2: Fetch the year-based match counts and aggregate
    german_data = aggregate_ngram_data(german_ids, 'ger')
    english_data = aggregate_ngram_data(english_ids, 'eng')
    russian_data = aggregate_ngram_data(russian_ids, 'rus')

    # Filter for years after 1950
    german_data = german_data[german_data.index >= 1950]
    english_data = english_data[english_data.index >= 1950]
    russian_data = russian_data[russian_data.index >= 1950]

    print(german_data)
    print(english_data)
    print(russian_data)

except Exception as e:
    print(e)
```

```python
import matplotlib.pyplot as plt

def plot_ngram_frequency(data, title):
    if not data.empty:
        plt.figure(figsize=(7, 4)) 
        data['absMatchCount'].plot(kind='line', marker='o')
        plt.title(title)
        plt.xlabel('Year')
        plt.ylabel('Absolute Match Count')
        plt.grid(True)
        plt.show() 
    else:
        print(f"No data to plot for: {title}")

try:
    german_ids = search_ngram("Internet", 'ger', 'cs')
    english_ids = search_ngram("Internet", 'eng', 'cs')
    russian_ids = search_ngram("Интернет", 'rus', 'cs')

    german_data = aggregate_ngram_data(german_ids, 'ger')
    english_data = aggregate_ngram_data(english_ids, 'eng')
    russian_data = aggregate_ngram_data(russian_ids, 'rus')

    german_data = german_data[german_data.index >= 1990]
    english_data = english_data[english_data.index >= 1990]
    russian_data = russian_data[russian_data.index >= 1990]

    plot_ngram_frequency(german_data, "Frequency of 'Internet' in German Corpus (1990-2019)")
    plot_ngram_frequency(english_data, "Frequency of 'Internet' in English Corpus (1990-2019)")
    plot_ngram_frequency(russian_data, "Frequency of 'Интернет' in Russian Corpus (1990-2019)")

except Exception as e:
    print(e)
```

```python
import requests
import pandas as pd
import plotly.express as px

def search_ngram(query, corpus='eng', flags='cs'):
    url = f"https://api.ngrams.dev/{corpus}/search"
    params = {
        'query': query,
        'flags': flags
    }
    response = requests.get(url, params=params)
    if response.status_code != 200:
        raise Exception(f"Error fetching data: {response.status_code} - {response.text}")
    data = response.json()
    ngram_ids = [ngram['id'] for ngram in data.get('ngrams', [])]
    return ngram_ids

def fetch_ngram_by_id(ngram_id, corpus='eng'):
    url = f"https://api.ngrams.dev/{corpus}/{ngram_id}"
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Error fetching ngram details: {response.status_code} - {response.text}")
    return response.json()

def aggregate_ngram_data(ngram_ids, corpus):
    all_data = pd.DataFrame()
    for ngram_id in ngram_ids:
        details = fetch_ngram_by_id(ngram_id, corpus)
        if 'stats' in details:
            df = pd.DataFrame(details['stats'])
            df['year'] = df['year'].astype(int)
            df.set_index('year', inplace=True)
            all_data = all_data.add(df, fill_value=0) if not all_data.empty else df
    return all_data

try:
    terms = {'Computer': ['Computer', 'Computer', 'Компьютер'], 'Internet': ['Internet', 'Internet', 'Интернет']}
    corpora = ['eng', 'ger', 'rus']
    
    data_dict = {}
    
    for term, queries in terms.items():
        for lang, corpus in zip(queries, corpora):
            ids = search_ngram(lang, corpus)
            data = aggregate_ngram_data(ids, corpus)
            data = data[data.index >= 1950]  
            data_dict[f"{term}_{corpus}"] = data

    year = 2019

    combined_data = {
        'Term': [],
        'Language': [],
        'Usage': []
    }
    
    for term, queries in terms.items():
        for corpus, lang in zip(corpora, queries):
            data_key = f"{term}_{corpus}"
            if year in data_dict[data_key].index:
                combined_data['Term'].append(term)
                combined_data['Language'].append(corpus.capitalize())
                combined_data['Usage'].append(data_dict[data_key].loc[year, 'absMatchCount'])

    df_combined = pd.DataFrame(combined_data)

    fig = px.treemap(df_combined, path=['Term', 'Language'], values='Usage',
                     title=f'Treemap Comparison of "Computer" and "Internet" Usage in {year}')
    fig.show()
    
except Exception as e:
    print(e)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.distance import cdist

cities = ['Amsterdam', 'Utrecht', 'Beijing', 'Copenhagen', 'Aarhus', 'Malmö', 'Eindhoven', 'The Hague', 'Osaka', 'Shanghai', 'Bremen', 'Dresden']
cycling = [40, 34, 32, 30, 27, 25, 24, 22, 21, 20, 19, 17]

sizes = [value * 1090 for value in cycling]

x_positions = np.random.rand(len(cities)) * 5
y_positions = np.random.rand(len(cities)) * 5

def adjust_positions(x_positions, y_positions, sizes):
    for i in range(len(x_positions)):
        for j in range(i+1, len(x_positions)):
            dist = np.sqrt((x_positions[i] - x_positions[j])**2 + (y_positions[i] - y_positions[j])**2)
            min_dist = np.sqrt(sizes[i]/np.pi) + np.sqrt(sizes[j]/np.pi)

            if dist < min_dist:
                angle = np.arctan2(y_positions[j] - y_positions[i], x_positions[j] - x_positions[i])
                overlap = min_dist - dist
                x_positions[j] += np.cos(angle) * overlap
                y_positions[j] += np.sin(angle) * overlap

    return x_positions, y_positions

for _ in range(100): 
    x_positions, y_positions = adjust_positions(x_positions, y_positions, sizes)

plt.figure(figsize=(13, 10))
plt.style.use('dark_background')

plt.scatter(x_positions, y_positions, s=sizes, alpha=0.8, color='#00FFBF', edgecolor='black')

for i, city in enumerate(cities):
    plt.text(x_positions[i], y_positions[i], f'{city}\n{cycling[i]}%', ha='center', va='center', fontsize=16, color='black', weight='bold')

plt.title('Cycling to Work: Cities with Most Bicycle Commuters', fontsize=18, color='#00FFFF')
plt.axis('off')  
plt.show()
```

```python
data_url_bli = 'https://sdmx.oecd.org/archive/rest/data/OECD,DF_BLI,/JPN+CAN.WL_TNOW+SW_LIFS+EQ_AIRP+HS_LE+SC_SNTWS+JE_EMPL..TOT?dimensionAtObservation=AllDimensions'

headers = {'Accept': 'application/vnd.sdmx.data+json; charset=utf-8; version=1.0'}

response = requests.get(data_url_bli, headers=headers)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: Received status code {response.status_code} and message: {response.text}")
```

```python
locations = data['data']['structure']['dimensions']['observation'][0]['values']
indicators = data['data']['structure']['dimensions']['observation'][1]['values']
measures = data['data']['structure']['dimensions']['observation'][2]['values']
inequalities = data['data']['structure']['dimensions']['observation'][3]['values']

location_dict = {str(i): loc['name'] for i, loc in enumerate(locations)}
indicator_dict = {str(i): ind['name'] for i, ind in enumerate(indicators)}
measure_dict = {str(i): mea['name'] for i, mea in enumerate(measures)}
inequality_dict = {str(i): ine['name'] for i, ine in enumerate(inequalities)}

observations = data['data']['dataSets'][0]['observations']

extracted_data = []

for obs_key, obs_value in observations.items():
    loc_idx, ind_idx, mea_idx, ine_idx = obs_key.split(':')
    country = location_dict[loc_idx]
    indicator = indicator_dict[ind_idx]
    measure = measure_dict[mea_idx]
    inequality = inequality_dict[ine_idx]
    value = obs_value[0] 
    extracted_data.append({
        'Country': country,
        'Indicator': indicator,
        'Measure': measure,
        'Inequality': inequality,
        'Value': value
    })

for item in extracted_data:
    print(item)
```

```python
observations = data['data']['dataSets'][0]['observations']

countries = data['data']['structures'][0]['dimensions']['observation'][0]['values']
indicators = data['data']['structures'][0]['dimensions']['observation'][1]['values']

country_mapping = {i['id']: i['name'] for i in countries}
indicator_mapping = {i['id']: i['name'] for i in indicators}

data_list = []

for key, value in observations.items():
    country_idx, indicator_idx, *_ = key.split(':')
    country_name = country_mapping.get(int(country_idx), 'Unknown Country')
    indicator_name = indicator_mapping.get(int(indicator_idx), 'Unknown Indicator')
    
    observation_value = value[0] 
    
    data_list.append({
        'Country': country_name,
        'Indicator': indicator_name,
        'Value': observation_value
    })

bli_df = pd.DataFrame(data_list)
bli_df.head()
```

```python
country_idx_map = {0: 'Japan', 1: 'Canada'}
indicator_idx_map = {idx: name for idx, name in enumerate(indicator_mapping.values())}

data_list_final = []

for key, value in observations.items():
    country_idx, indicator_idx, *_ = map(int, key.split(':'))

    country_name = country_idx_map.get(country_idx, 'Unknown Country')
    indicator_name = indicator_idx_map.get(indicator_idx, 'Unknown Indicator')
    
    observation_value = value[0]
    
    data_list_final.append({
        'Country': country_name,
        'Indicator': indicator_name,
        'Value': observation_value
    })

bli_df_final = pd.DataFrame(data_list_final)
bli_df_final.head()
```

```python
gdp_api_url = 'https://sdmx.oecd.org/public/rest/data/OECD.CFE.EDS,DSD_REG_ECO@DF_GDP,2.0/A..JPN+CAN..GDP..Q.USD_PPP_PS?startPeriod=2018&dimensionAtObservation=AllDimensions'
response = requests.get(gdp_api_url, headers={'Accept': 'application/vnd.sdmx.data+json; charset=utf-8'})

if response.status_code == 200:
    gdp_data = response.json()

    with open('gdp_data_japan_canada.json', 'w') as f:
        json.dump(gdp_data, f, indent=4)

    print(json.dumps(gdp_data, indent=4))
else:
    print(f"Error: Received status code {response.status_code}")
```

```python
gdp_data = gdp_data['data']['dataSets'][0]['observations']

country_map = {0: 'Japan', 1: 'Canada'}

gdp_values_extracted = {}

for key, value in gdp_data.items():
    if ':' in key:
        country_idx = int(key.split(':')[2]) 
        country_name = country_map.get(country_idx, 'Unknown')
        gdp_value = value[0]  
        
        gdp_values_extracted[country_name] = gdp_value

gdp_values_extracted
```

```python
life_satisfaction_data = bli_df_final[bli_df_final['Indicator'] == 'Life satisfaction']
life_satisfaction_values = life_satisfaction_data.set_index('Country')['Value'].to_dict()

def standardize(values):
    min_val = min(values.values())
    max_val = max(values.values())
    return {k: (v - min_val) / (max_val - min_val) for k, v in values.items()}

gdp_standardized = standardize(gdp_values_extracted)
life_satisfaction_standardized = standardize(life_satisfaction_values)

comparison_df = pd.DataFrame({
    'Country': ['Japan', 'Canada'],
    'GDP (Standardized)': [gdp_standardized['Japan'], gdp_standardized['Canada']],
    'Life Satisfaction (Standardized)': [life_satisfaction_standardized['Japan'], life_satisfaction_standardized['Canada']]
})

print(comparison_df)
```
